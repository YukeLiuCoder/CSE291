Q3: Implement steepest descent with learning rate 0.1 and 0.01. Plot how the function valueschange in the first 100 iterations

```python
import matplotlib.pyplot as plt
cur_x = 1 # The algorithm starts at x=1
rate = 0.01 # Learning rate
iters = 0 #iteration counter
df = lambda x: 4*(x**3) #Gradient of our function 
```

```python
#Implement gradient descent 

x_list=[]
it_list=[]
y_list=[]
while iters < 100:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - rate * df(prev_x) #Grad descent
    x_list.append(cur_x)
    cur_y = cur_x**4
    y_list.append(cur_y)
    previous_step_size = abs(cur_x - prev_x) #Change in x
    it_list.append(iters)
    iters = iters+1 #iteration count
plt.plot(it_list,y_list)
#it_list
```

Q5: Implement the Newton direction. Compare step size 1 with two other choices of step sizes ofyour choice. Plot the first 100 iterations

```python
import matplotlib.pyplot as plt
cur_x = 1 # The algorithm starts at x=1
rate = 0.1 # Learning rate
iters = 0 #iteration counter
df = lambda x: 4*(x**3) #Gradient of our function 
d2f = lambda x: 12*(x)**2
#Implement gradient descent 

x_list=[]
it_list=[]
y_list=[]

while iters < 100 :
    prev_x = cur_x #Store current x value in prev_x
    d = - (d2f(cur_x))**(-1) *df(cur_x)
    
    cur_x = cur_x + rate * d #Grad descent
    x_list.append(cur_x)
    cur_y = cur_x**4
    y_list.append(cur_y)
    it_list.append(iters)
    iters = iters+1 #iteration count
plt.plot(it_list,y_list)
#it_list
```
